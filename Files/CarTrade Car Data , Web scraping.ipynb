## Final data frame

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup

final = pd.DataFrame()

for j in range(1,100):
    url= "https://www.cartrade.com/buy-used-cars/delhi/c/p-{}".format(j)
    
    headers = {'User-Agend': 'Mozilla/5.0 (Windows NT 6.3; Win64 x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers = headers).text

    soup = BeautifulSoup(webpage , 'lxml')
    
    name = []
    price = []
    km = []
    car_type = []
    location = []
    
    # name
    car = (soup.find_all('h2',class_='h2heading truncate'))
    for i in car:
        name.append(i.find('a').text.strip())
        
    #price
    p = (soup.find_all('div',class_='cr_prc'))
    for i in p:
        price.append(i.text[1:9])    
        
    price.pop()
    # km
    k = (soup.find_all('div',class_='info_cr_new'))
    for i in k:
        km.append(i.find_all('li')[0].text.strip())   
            
    # type
    t = (soup.find_all('div',class_='info_cr_new'))
    for i in t:
        car_type.append(i.find_all('li')[2].text)   
        
    #location
    l = (soup.find_all('div',class_='info_cr_new'))
    for i in l:
        location.append(i.find_all('li')[4].text.strip())  
        
            
    data = {'Car_name':name , 'KM':km , 'Type':car_type , 'Location': location , 'Price' : price}
    
    #print(len(name),len(km),len(car_type),len(location),len(price))

    df = pd.DataFrame(data)
    final = final.append(df,ignore_index=True)
```
